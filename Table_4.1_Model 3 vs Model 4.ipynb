```python
import math
import matplotlib.pyplot as plt
import random
import numpy
import math
import matplotlib.pyplot as plt
import random
from k_means1 import k_means
from scipy import signal
from scipy.spatial.distance import pdist,squareform
from sklearn import cluster
```

```python
from __future__ import print_function
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage,fcluster
import matplotlib
import matplotlib.pyplot as plt
get_ipython().magic(u'matplotlib inline')
from sklearn.cluster import KMeans
from collections import Counter
import datetime

def accu(label, value):
    count1 = 0
    count2 = 0
    for h in range(len(label)):
        for g in range(len(label[0])):
            if h > (n-2*m)/2-1 and label[h][g] ==1:
                count1 += 1
            if h <(n-2*m)/2 and label[h][g] ==1:
                count2 += 1
    return max(abs(count1-value)+count2, abs(count2-value)+count1)/(value*2)

n = 128
m = round(n**0.5)
r = round(n**0.5)
mat1 =  np.array([float(0) for i in range((2*r+1)**2)]).reshape(2*r+1, 2*r+1)
mat2 =  np.array([float(0) for i in range((2*r+1)**2)]).reshape(2*r+1, 2*r+1)
for v in range(2*r+1):
    mat1[r][v] = 1/(2*r+1)
    mat2[v][r] = 1/(2*r+1)

result_s = []
result_ps = []
result_k = []
for rr in range(50):
    print(rr)
    start_time = datetime.datetime.now() 
    nn= int((n+2*r)*(n+2*r))
    data1 = np.random.normal(0, 1, nn).reshape(int(n+2*r), int(n+2*r))
    data2 = np.random.normal(0, 1, nn).reshape(int(n+2*r), int(n+2*r))
    data = data1[:n,:n]
    data12 = data2[:n,:n]
    for g in range(n):
        for h in range(n):
            da1 = data1[g:g+2*r+1,h:h+2*r+1]
            da2 = data2[g:g+2*r+1,h:h+2*r+1]
            data[g][h] = sum(sum(np.multiply(mat1, da1)))
            data12[g][h] = sum(sum(np.multiply(mat2, da2)))

    data = data - numpy.mean(data)
    data = data/numpy.std(data)
    data12 = data12 - numpy.mean(data12)
    data12 = data12/numpy.std(data12)

    for h in range(n):
        for g in range(n):
            if h > n/2-1:
                data[h][g]=data12[h][g]
    plt.imshow(data, cmap='gray')
    plt.show()

    lis = []
    for i in range(m, n-m):
        for j in range(m, n-m):
            mat = data[i-m:i+m+1,j-m:j+m+1]
            matt = numpy.rot90(mat, 2)/((2*m+1)**2)
            res = signal.convolve2d(mat, matt, mode='full')
            res = res[m:3*m+1,m:3*m+1]
            res = res.flatten().tolist()
            res.append(3*i/n)
            res.append(3*j/n)
            lis.append(res)        
    

    Z = linkage(lis, 'single')
    ZZ = Z[:,3]
    y= 0
    while ZZ[y] <= ((n-2*m)**2)*0.6:
        y=y+1
    y=y-1
    label = fcluster(Z,Z[y,2],'distance')
    print(max(Z[0:y,3])/((n-2*m)**2))

    val = Counter(label).most_common(1)[0][0]
    label = [0 if x!=val else 1 for x in label]
    label1 = np.array(label).reshape(n-2*m,n-2*m)
    value = (n-2*m)**2/2
    accuracy0 = accu(label1,value)
    print(accuracy0, sum(sum(label1))/(value*2))

    plt.imshow(label1, cmap='gray')
    plt.show()


    clst=cluster.AgglomerativeClustering(n_clusters=2,linkage='ward')
    label11=clst.fit_predict(lis).reshape(n-2*m,n-2*m)
    value = (n-2*m)**2/2
    accuracy1 = accu(label11,value)
    print(accuracy1, sum(sum(label11))/(value*2))
    plt.imshow(label11, cmap='gray')
    plt.show()


    Clu1, label1 = k_means(lis, 2, 10)
    label11 = np.array(label1).reshape(n-2*m,n-2*m)
    value = (n-2*m)**2/2
    accuracy2 = accu(label11,value)
    print(accuracy2, sum(sum(label11))/(value*2))
    plt.imshow(label11, cmap='gray')
    plt.show()
    
    result_s.append(accuracy0)
    result_ps.append(accuracy1)
    result_k.append(accuracy2)
    end_time = datetime.datetime.now()     
    interval = (end_time-start_time).seconds
    final_time = interval/60.0 
    print ('Time:\t', final_time)
    
print('Single:',np.mean(result_s))
print('Ward:',np.mean(result_ps))
print('K-means:',np.mean(result_k))
```
